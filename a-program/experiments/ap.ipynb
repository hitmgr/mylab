```python

import flgo.experiment.analyzer as fea
import matplotlib.pyplot as plt
%matplotlib inline

# 设置更好看的默认样式
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.axisbelow'] = True

# 定义任务路径和划分方式
# task_prefix = '/home/mgr/sss/easyFL-FLGo/a-program/tasks/cifar10/my_and_fedavg_and_fedprox_10'
task_prefix = '/home/mgr/sss/easyFL-FLGo/a-program/tasks/timeTry'
# partitions = ['iid', 'div01', 'div05', 'div09', 'dir01', 'dir10', 'dir50']
partitions = ['iid']

# 创建一个字典来存储所有记录
all_records = {}

for partition in partitions:
    task_path = f"{task_prefix}/{partition}"
    records = fea.load_records(task_path, ['fedavg', 'fedprox', 'MyAlgorithm'])
    all_records[partition] = records

def plot_performance(records, partition):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12))
    colors = ['r', 'g', 'b']
    labels = ['FedAvg', 'FedProx', 'MyAlgorithm']
    
    for i, record in enumerate(records):
        rounds = record.log['communication_round']
        accuracy = record.log['test_accuracy']
        loss = record.log['test_loss']
        
        # Plot accuracy
        ax1.plot(rounds, accuracy, color=colors[i], label=f'{labels[i]} - Accuracy')
        
        # Plot loss
        ax2.plot(rounds, loss, color=colors[i], linestyle='--', label=f'{labels[i]} - Loss')
    
    ax1.set_title(f'Test Accuracy ({partition})')
    ax1.set_xlabel('Rounds')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    
    ax2.set_title(f'Test Loss ({partition})')
    ax2.set_xlabel('Rounds')
    ax2.set_ylabel('Loss')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

# 为每个划分方式绘制图表
for partition, records in all_records.items():
    plot_performance(records, partition)
```

# 表格

```python
import flgo.experiment.analyzer as fea
from prettytable import PrettyTable

# 定义任务路径和划分方式（保持与您的代码一致）
task_prefix = '/home/mgr/sss/easyFL-FLGo/a-program/tasks/cifar10/100clients_change_patience/adjustRate_0.005'
partitions = ['iid', 'div01', 'div05', 'div09', 'dir01', 'dir10', 'dir50']

# 创建一个字典来存储所有记录（保持与您的代码一致）
all_records = {}

for partition in partitions:
    task_path = f"{task_prefix}/{partition}"
    records = fea.load_records(task_path, ['fedavg', 'fedprox', 'MyAlgorithm'])
    all_records[partition] = records
    
# 创建表格
table = PrettyTable()
table.field_names = ["Partition", "Algorithm", "Final Test Accuracy", "Lowest Test Loss"]
table.align = "l"  # 左对齐所有列

# 设置表格样式
table.border = True
table.header = True

# 定义获取最终测试准确率的函数
def get_final_accuracy(record):
    return record.log['test_accuracy'][-1]

# 定义获取最低测试损失的函数
def get_lowest_loss(record):
    return min(record.log['test_loss'])

# 填充表格
for partition, records in all_records.items():
    for j, record in enumerate(records):
        algorithm = record.algorithm
        final_accuracy = get_final_accuracy(record)
        lowest_loss = get_lowest_loss(record)
        
        table.add_row([
            partition if j == 0 else "",  # 只在每个partition的第一行显示partition名称
            algorithm,
            f"{final_accuracy:.4f}",
            f"{lowest_loss:.4f}"
        ])
    
    # 在每个partition之后添加分隔行
    if partition != partitions[-1]:  # 不在最后一个partition后添加分隔行
        table.add_row(["-" * 20] * 4)

# 打印表格
print(table)
```
